#### Cleaning the Survey Datasets 

```python
# fliter each year 
df_2022 = df_2022.filter(['Employment', 'YearsCode','EdLevel','Country', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith', 'DevType', 'ConvertedCompYearly'])
df_2021 = df_2021.filter(['Employment', 'YearsCode','EdLevel','Country', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith','DevType', 'ConvertedCompYearly'])
df_2020 = df_2020.filter(['Employment', 'YearsCode','EdLevel','Country',  'LanguageWorkedWith', 'LanguageDesireNextYear','DevType', 'CompTotal'])
df_2019 = df_2019.filter(['Employment', 'YearsCode','EdLevel','Country',  'LanguageWorkedWith', 'LanguageDesireNextYear','DevType','CompTotal'])

# Renaming 2020 and 2019 to match previous data set
df_2020.rename(columns={'LanguageWorkedWith': 'LanguageHaveWorkedWith', 'LanguageDesireNextYear': 'LanguageWantToWorkWith', 'CompTotal': 'ConvertedCompYearly'}, inplace=True)
df_2019.rename(columns={'LanguageWorkedWith': 'LanguageHaveWorkedWith', 'LanguageDesireNextYear': 'LanguageWantToWorkWith', 'CompTotal' :'ConvertedCompYearly'}, inplace=True)

# Concatenates the data sets
dataYears = pd.concat([df_2019, df_2020, df_2021, df_2022])

# Removes any rows with NaN as an answer 
dataYears = dataYears.dropna()

# Prints cleaned data set for 2019-2022 
print(dataYears)
```

#### Visualizing the Trend of Popularity of Programming Languages Since 1988 Based on Tiobe Index 

```python
df =Popularity_current

# Calculate the sum of ratings for all programming languages
total_ratings = df['Ratings'].sum()

# Create a new column with the programming language names and ratings below 7% labeled as "Other"
df['Language'] = df.apply(lambda x: x['Programming Language'] if x['Ratings'] >= 0.07 else 'Other', axis=1)

# Calculate the sum of ratings for the programming languages labeled as "Other"
other_ratings = df[df['Language'] == 'Other']['Ratings'].sum()

# Remove the rows with ratings below 7%
df = df[df['Ratings'] >= 0.07]

# Add the "Other" row with the sum of ratings below 7%
other_row = {'Programming Language': 'Other', 'Ratings': other_ratings, 'Language': 'Other'}
df = pd.concat([df, pd.DataFrame(other_row, index=[0])], ignore_index=True)

# Sort the dataframe by Ratings in descending order
df = df.sort_values(by='Ratings', ascending=False)

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(df['Ratings'], labels=df['Language'], autopct='%1.1f%%')
ax.set_title('Popularity of Programming Languages (March 23)')
plt.show()
```

#### Visualizing the Relationship between programming languages and education levels based on the StackoverFlow Annual Survey 

```python
data = dataYears

# Split the values in the LanguageHaveWorkedWith column into separate rows
data = data.explode('LanguageHaveWorkedWith')

# Group the data by education level and programming language
grouped_data = data.groupby(['EdLevel', 'LanguageHaveWorkedWith']).size().reset_index(name='count')

# Pivot the data to create a matrix with education levels as rows and programming languages as columns
pivot_data = grouped_data.pivot(index='EdLevel', columns='LanguageHaveWorkedWith', values='count').fillna(0)

# Create the stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8))

# Add labels and title
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.title('Programming Languages Used by Education Level')
plt.legend(ncol=len(df.columns),bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```

#### Visualizing the Relationship between programming languages and countries, based on the StackoverFlow Annual Survey 

```python
df_copy = df.copy()
df_copy['LanguageHaveWorkedWith'] = df_copy['LanguageHaveWorkedWith'].apply(lambda x: ';'.join(map(str, x)) if isinstance(x, (list, tuple)) else str(x))

# Get the top 20 countries by count
top_countries = df['Country'].value_counts().head(20)

# Get the programming languages used in each country
languages_by_country = df.groupby('Country')['LanguageHaveWorkedWith'].apply(lambda x: ';'.join(map(str, x))).reset_index()

# Concatenate the two dataframes based on the 'Country' column
merged = pd.merge(languages_by_country, top_countries, left_on='Country', right_index=True).reset_index(drop=True)

# Get the top programming languages overall
all_languages = pd.Series(';'.join(df['LanguageHaveWorkedWith'].astype(str)).split(';'))
top_languages = all_languages.value_counts().head(20).index.tolist()

# Create a dictionary to store the counts of each language
language_counts = {language: [] for language in top_languages}

# Create a dictionary to map each language to its color
language_colors = {language: f'C{i}' for i, language in enumerate(top_languages)}

# Iterate over the merged dataframe to count the number of occurrences of each language
for _, row in merged.iterrows():
    country_languages = row['LanguageHaveWorkedWith'].split(';')
    for language in top_languages:
        count = country_languages.count(language)
        language_counts[language].append(count)

# Create the chart
fig, ax = plt.subplots(figsize=(12, 8))

# Iterate over the languages and add each language and its corresponding color to the legend
for language in top_languages:
    ax.barh(merged['Country'], language_counts[language], height=0.8, color=language_colors[language], alpha=0.7)

ax.legend(labels=top_languages, loc='lower right')
ax.set_xlabel('Number of programmers')
ax.set_ylabel('Country')
ax.set_title('Programming languages used by Country')
plt.show()
```

#### Visualization of the Relationship between programming languages and employment status based on the StackoverFlow Annual Survey 

```python
df = dataYears

# Replace the values in the 'Employment' column
df.loc[:, 'Employment'] = df['Employment'].replace({
    'Employed, full-time': 'Employed full-time',
    'Full-time employed':'Employed full-time',
    'Employed part-time': 'Part-time employed',
    'Not employed, and not looking for work': 'Other',
    'Independent contractor, freelancer, or self-employed': 'Private contractor'
})

# Calculate the percentage of each employment category
employment_counts = df['Employment'].value_counts(normalize=True) * 100

# Group values under 2% into the "other" category
other_percent = 2
mask = employment_counts < other_percent
mask = mask.reindex(employment_counts.index, fill_value=False)

other = employment_counts[mask].sum()
other_percent = round(other, 1)

employment_counts = employment_counts[~mask]
employment_counts['Other'] = other_percent

plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%')
plt.title('Employment Status of Respondents')
plt.axis('equal')
plt.show()
```

#### Visualization of Top Programming Languages of Full-Time Employed  based on the StackoverFlow Annual Survey 

```python
df = dataYears
# Filter the data to only include those who are employed full-time
df = df[df['Employment'] == 'Employed full-time']
df = dataYears
# Create a list of all programming languages mentioned in the dataset
all_languages = set()
for _, row in df.iterrows():
    if row["LanguageHaveWorkedWith"]:
        if isinstance(row["LanguageHaveWorkedWith"], list):
            all_languages.update(";".join(row["LanguageHaveWorkedWith"]).split(";"))
        else:
            all_languages.update(row["LanguageHaveWorkedWith"].split(";"))

# Create a dictionary that maps each programming language to its count among Employed (Full-Time)
language_counts = {lang: 0 for lang in all_languages}
for _, row in df[df["Employment"].str.contains("Employed full-time")].iterrows():
    if row["LanguageHaveWorkedWith"]:
        if isinstance(row["LanguageHaveWorkedWith"], list):
            for lang in ";".join(row["LanguageHaveWorkedWith"]).split(";"):
                language_counts[lang] += 1
        else:
            for lang in row["LanguageHaveWorkedWith"].split(";"):
                language_counts[lang] += 1

language_df = pd.DataFrame.from_dict(language_counts, orient="index", columns=["count"])
language_df = language_df.nlargest(15, "count")

# Create the bar chart
plt.bar(language_df.index, language_df["count"])
plt.xticks(rotation=90)
plt.xlabel("Programming language")
plt.ylabel("Count")
plt.title("Programming languages used by Full-Time Empolyed")
plt.show()
```

#### Percentage of Developer Types within the Workforce based on the StackoverFlow Annual Survey 

```python
df=dataYears

# Count the number of occurrences for each DevType
devtype_counts = df['DevType'].value_counts()

# Calculate the percentage of each DevType
devtype_percents = devtype_counts * 100 / devtype_counts.sum()

# Create a mask for DevTypes with less than 5% occurrence
mask = devtype_percents < 3

# Group the DevTypes with less than 5% occurrence as 'Other'
devtype_percents_other = devtype_percents[mask].sum()
devtype_percents = devtype_percents[~mask]
devtype_percents['Other'] = devtype_percents_other

# Create the pie chart
plt.pie(devtype_percents, labels=devtype_percents.index, autopct='%1.1f%%')
plt.title('Percentage of DevType')
plt.show()
```

#### Top Programming Langauges based on Total Yearly Salary based on the StackoverFlow Annual Survey 

```python
df = dataYears
# Explode the 'LanguageHaveWorkedWith' column so that each language is in a separate row
df_languages = df.explode('LanguageHaveWorkedWith')

# Drop rows with missing salary data
df_languages = df_languages.dropna(subset=['ConvertedCompYearly'])

# Group the data by programming language and calculate the median salary for each language
df_languages = df_languages[['LanguageHaveWorkedWith', 'ConvertedCompYearly']].groupby('LanguageHaveWorkedWith').median().sort_values(by='ConvertedCompYearly', ascending=False)

# Create a bar chart showing the median salary for each programming language
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(df_languages.index, df_languages['ConvertedCompYearly'])
ax.set_xticks(range(len(df_languages.index)))
ax.set_xticklabels(df_languages.index, rotation=90)
ax.set_xlabel('Programming Language')
ax.set_ylabel('Median Annual Salary (USD)')
ax.set_title('Median Annual Salary by Programming Language')
plt.show()
```
